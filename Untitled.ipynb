```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
college = pd.read_csv('College.csv')
```

```python
college = college.rename(columns = {'Unnamed: 0':'School'})
```

```python
college = college.set_index('School')
```

```python
college[['Private']]
```

```python
college.describe()
```

```python
import seaborn as sns
```

```python
df1 = college.iloc[:,:10]
df1
```

```python
sns.pairplot(df1)
plt.show()
plt.savefig('pairplot')
```

```python
plt.figure(figsize = (10,8))
sns.boxplot(data = college, x = 'Private', y = 'Outstate')
plt.show()
```

```python
college.loc[college['Top10perc'] > 50, 'Elite'] = 'Yes'
college.loc[college['Top10perc'] < 50, 'Elite'] = 'No'
college
```

```python
plt.figure(figsize = (10,8))
sns.boxplot(data = college, x = 'Elite', y = 'Outstate')
plt.show()
```
